## GRAPH TWEETS CONTRIES

```python
tweets_colors = ['red', 'orangered', 'darkorange', 'goldenrod', 'gold', 'yellow', 'greenyellow', 'lawngreen', 'seagreen', 'turquoise', 'deepskyblue', 'royalblue', 'slateblue', 'indigo', 'deeppink']
```

```python
fig = matplotlib.pyplot.gcf()
fig.set_size_inches(16, 4.5)

fig = plt.gcf()
fig.set_size_inches(16, 4.5)


sq.plot(sizes = auxc['target_bool'], label = auxc['country'], alpha=.6, color = tweets_colors)
plt.axis('off')
plt.title("Top 15 - PaÃ­ses con mas tweets ",fontsize=23,fontweight="bold")
plt.show()
```

```python
countrytweets = tweets.groupby(['country'], as_index = False ).agg({'target_bool': 'count'})
countrytweets.sort_values(by=['target_bool'], inplace=True, ascending=False )
TweetsInContries = countrytweets.head(15)
TweetsInContries
```

```python
poblaciones = ['328200000', '195900000', '37590000', '1095351995', '24990000', '267700000', '66650000', '51390000', '4900000','212200000' , '83020000', '106700000', '46940000', '126500000', '60360000']
TweetsInContries['poblacion'] = poblaciones
TweetsInContries
```

```python
fig = matplotlib.pyplot.gcf()
fig.set_size_inches(20, 4.5)

fig = plt.gcf()
fig.set_size_inches(20, 4.5)


# create figure and axis objects with subplots()
fig,ax = plt.subplots()
# make a plot
ax.plot( TweetsInContries['country'], TweetsInContries['target_bool'], color="red", marker="o" )
# set x-axis label
ax.set_xlabel("Countries",fontsize=10)
# set y-axis label
ax.set_ylabel("Tweets with their location",color="red",fontsize=14)



# twin object for two different y-axis on the sample plot
ax2=ax.twinx()
# make a plot with different y-axis using second axis object
ax2.plot(TweetsInContries['country'],TweetsInContries['poblacion'] ,color="blue",marker="o")
ax2.set_ylabel("Poblacion en los paises",color="blue",fontsize=14)
plt.show()

# save the plot as a file
fig.savefig('two_different_y_axis_for_single_python_plot_with_twinx.jpg',
            format='jpeg',
            dpi=100,
            bbox_inches='tight')
```
