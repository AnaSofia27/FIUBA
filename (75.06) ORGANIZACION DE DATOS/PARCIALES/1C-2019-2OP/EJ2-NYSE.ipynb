### Por otro lado se cuenta con el archivo s&p500.csv de formato (symbol, company_name) que indica aquellas acciones que deben ser consideradas para calcular el índice Standard & Poor’s 500 (S&P 500). Se pide calcular el valor diario del índice S&P 500, teniendo en cuenta que el mismo se calcula como el promedio del valor promedio de las mediciones que tuvo cada acción ese día (open, measure_midday, measure_afternoon, close), para las 500 acciones que se encuentran en el archivo s&p500.csv. El resultado debe estar en un dataframe de la forma:
### (date, index_name, value).
#### Por ejemplo, una entrada del mismo sería (‘2019-03-24’, ‘SP500’, ‘35.46’)

```
#Me quedo con las acciones importantes para S&P 
#(symbol, date, open, measure_midday, measure_afternoon, close, volume, company)
NyseSp = Sp.merge(Nyse, on='symbol',how='left')
```

```
#Calculo el promedio diario para cada accion cada dia (por fila)
#(symbol, date, open, measure_midday, measure_afternoon, close, volume, company, daily_mean)
NyseSp['daily_mean'] = NyseSp.groupby('symbol', 'date').apply(lambda x: (x.open + x.measure_midday + x.measure_afternoon +x.close) / 4 ) 
```

```
#Tengo que calcular el promedio de los promedios diarios
Sp500 = NyseSp.groupby('date').agg({'daily_mean': 'mean' }).reset_index()
```

```
#Agrego la columna que me piden
Sp500['index_name'] = "SP500"
```

```
#Me piden el DataFrame con el formato : (date, index_name, value).
Sp500.columns = ['date', 'value', 'index_name']
```
