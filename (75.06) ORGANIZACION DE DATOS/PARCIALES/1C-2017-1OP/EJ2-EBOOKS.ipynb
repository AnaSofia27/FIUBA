###b) Obtener un DataFrame que indique qué libros tienen una diferencia de rating promedio mayor al 20% entre el sitio de Ebooks y GoodReads


```
#Tenemos el df ReviewsEbooks (user_id, book_id, rating, timestamp).
#Tenemos el df GoodReads (book_id, book_name, avg_rating).
```

```
#Distintos usuarios pudieron calificar el mismo libro asi que agrupo por libro y calculo el promedio.
EBooksMeanRaiting = Reviews.gropuby(['book_id']).agg({'rating': 'mean'}).reset_index()

#Renombro mis columnas.
EBooksMeanRaiting.columns = ['Book_id', 'MeanRating_Ebooks']
GoodReads.columns = ['Book_id', 'Book_name','MeanRating_GoodReads']
```

```
#Para Juntar dos DataFrames les pongo el mismo index.
EBooksMeanRaiting.set_index('Book_id')
GoodReads.set_index('Book_id')

#Junto ambos DataFrames.
AmbasPaginas = EBooksMeanRaiting.join(GoodReads)
```

```
#Creo una columna con la diferencia de ratings
AmbasPaginas['DiferenciaRatings'] = abs( AmbasPaginas['MeanRating_Ebooks'] -  AmbasPaginas['MeanRating_GoodReads'] )

#Filtro las filas que cumplen con diferencia de rating promedio mayor al 20% entre el sitio de Ebooks y GoodReads.
#Para esto suponemos que el rating va de 0-100 en ambos casos.
FiltroMayor20 = AmbasPaginas['DiferenciaRatings'] > 20
DiferenciaMayorA20 = AmbasPaginas[FiltroMayor20]
```
