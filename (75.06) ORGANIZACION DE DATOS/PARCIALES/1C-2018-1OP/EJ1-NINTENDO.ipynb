### 1) Nintendo of America (EEUU) tiene información de ventas de videojuegos físicas mensuales totalizadas en EEUU las cuales se realizan en cadenas de tiendas de videojuegos en el siguiente RDD:

### VENTAS (id_videojuego, id_tienda, mes, anio, total_ventas_mensuales).

### Por otro lado tenemos un RDD con información de las tiendas y de su ubicación:
###  TIENDAS (id_tienda, direccion, latitud,longitud, codigo_postal, estado).

### Con esta información escribir un programa en pySpark para obtener la tienda que realizó menor cantidad de ventas en el estado de “Georgia” en todo 2017.

```
#Filtro por año 2017.
Ventas = Ventas.filter(lambda x: x[3] == 2017)
```

```
#Filtro por estado de Georgia.
Tiendas = Tiendas.filter(lambda x: x[5] == 'Georgia')
```

```
#Arreglo mi RDD Ventas para poder juntar con el RDD Tiendas.
#(id_tienda, total_ventas_mensuales)
Ventas = Ventas.map(lambda x: (x[1], x[4]) )
```

```
#Arreglo mi RDD Tiendas para poder juntar con el RDD Ventas.
#(id_tienda)
Tiendas = Tiendas.map(lambda x: x[0] )
```

```
#Junto ambos RDDs --> (id tienda, total_ventas_mensuales).
VentasEnGeorgia = Tiendas.OuterLeftjoin(Ventas)
```

```
#Sumo la cantidad total de ventas por tienda.
#(id tienda, total_ventas).
VentasEnGeorgia = VentasEnGeorgia.reduceByKey(lambda a,b: a+b)
```

```
#Me quedo con la que menos ventas tuvo
#(id tienda, total_ventas).
MenorVentasEnGeorgia = VentasEnGeorgia.reduce(lambda a,b : a if a[1] < b[1] else b )
```
