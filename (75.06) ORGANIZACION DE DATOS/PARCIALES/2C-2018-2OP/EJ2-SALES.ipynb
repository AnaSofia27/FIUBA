(por ejemplo si el promedio de Abril de "Men" es 120 dolares y un producto se vendió en Abril a 135 dolares lo tenemos que listar). Usar Transform.

```
#Filtro las venta de productos con categoria 'Men'
FiltroMen = VentaProductos['Categoria'] == 'Men'
VentaProductos = VentaProductos[FiltroMen]
```

```
#Junto ambos DataFrames (Dia, Mes, Año, ProductID, Importe, Promedio_Mensual, Titulo, Categoria, Descripcion )
#Notar que solo me van a quedar productos de la categoria 'Men'
VentaProductos = Sales.merge(Products, on='ProductID', how='left')
```

```
#Genero una nueva columna con los promedios mensuales de los productos de una misma categoria
#(Dia, Mes, Año, ProductID, Importe, Titulo, Categoria, Descripcion  Promedio_Mensual) .
#Notar que dice el promedio mensual DE LOS PRODUCTOS de la categoria, es decir no es el promedio del mismo producto.
Sales['Promedio_Mensual'] = Sales.groupby(['Mes', 'Año'])['Importe'].transform('mean')
```

```
#Filtro los Productos que: Importe > Promedio_Mensual
FiltroMayorImporte = VentaProductos['Importe'] > VentaProductos['Promedio_Mensual']
VentaProductos = VentaProductos[FiltroMayorImporte]
```

```
#Titulos de los productos de la categoria 'Men' en los cuales el Importe es mayo al Promedio_Mensual
Resultado = VentaProductos['Titulo']
```
