```
#Filtro por Octubre y los años 2016 y 2017
FiltroOctubre = (Encuesta['mes_nacimiento'] == 10)
FiltroAño = (Encuesta['anio_nacimiento'] == 2017) | (Encuesta['anio_nacimiento'] == 2016)
EncuestaFiltradas = Encuesta[FiltroOctubre & FiltroAño]
```

```
#Cuento el promedio de partos por cesarea
porc_cesarea = EncuestaFiltradas.groupby('anio_nacimiento')['tipo_parto'].agg(lambda x: (x==2).mean())
```
