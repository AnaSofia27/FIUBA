### Por otro lado se tienen otro RDD que tiene información de todos los videojuegos que se venden en su plataforma con el siguiente formato:
####(id_videojuego, titulo, rating_pegi, rating_esbr).
###Tener en cuenta que un mismo videojuego se puede vender en distintos países y esos nos permitirá obtener métricas a nivel global.

##A) Obtener el videojuego con más ventas digitales globales (es decir en todos los países) en un RDD con el siguiente formato: (id_videojuego, titulo, total), siendo total la cantidad total de ventas digitales globales

```
#Me quedo con los campos importantes del primer RDD (id_videojuego, ventas_diarias)
VentasNintendo = RDDVentas.map(lambda x: (x[0], x[4]) )
```

```
#Agrupo por videojuego y calculo total_ventas :  (id_videojuego, ventas_totales)
VentasTotales = VentasNintendo.reduceByKey(lambda a,b : a+b)
```

```
#Reordeno mi otro RDD para poder juntar y obtener la info (id_videojuego, titulo)
Informacion = RDDInformacion.map(lambda x: (x[0], x[1]) )
```

```
#Junto Ambos RDD (id_videojuego, (total_ventas, titulo))
VentasDigitales = VentasTotales.LeftOuterjoin(Informacion)
```

```
#Reordeno mi RDD para dejarlo de la forma pedida  (id_videojuego, titulo, total)
VentasDigitales = VentasDigitales.map(lambda x: (x[0], x[1][1], x[1][0]) ) 
```

```
#Busco el que tiene mas ventas
MasVendido = VentasDigitales.reduce(lambda a,b: a if a[2] > b[2] else b)
```

**LO QUE ESTA A CONTINUACION SERIA SI TENDRIA QUE TENER EN CUENTA EL PAIS!!!!**

```
#Me quedo con los campos importantes del primer RDD ((codigo_pais, id_videojuego), total_ventas_diarias )
VentasNintendo = RDDVentas.map(lambda x: ((x[1], x[0]), x[4]) ))
```

```
#Agrupo por pais y videojuego ya que un mismo pais pudo tener ventas en distintos dias 
#((codigo_pais, id_videojuego), total_ventas_diarias)
VentasTotalesNintendo = VentasNintendo.reduceByKey(lambda a b: a + b)
```

```
#Reordeno mi RDD para poder juntar con el otro RDD y obtener info (id_videojuego, total_ventas)
VentasTotalesNintendo = VentasTotalesNintendo.map(lambda x: (x[0][1], x[1]) )
```

```
#Reordeno mi otro RDD para poder juntar obtener info (id_videojuego, titulo)
Informacion = RDDInformacion.map(lambda x: (x[0], x[1]) )
```

```
#Junto Ambos RDD (id_videojuego, (total_ventas, titulo))
VentasDigitales = VentasTotalesNintendo.LeftOuterjoin(Informacion)
```

```
#Reordeno mi RDD para dejarlo de la forma pedida  (id_videojuego, titulo, total)
VentasDigitales = VentasDigitales.map(lambda x: (x[0], x[1][1], x[1][0]) ) 
```

```
#Busco el que tiene mas ventas
MasVendido = VentasDigitales.reduce(lambda a,b: a if a[2] > b[2] else b)
```
